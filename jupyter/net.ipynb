```python
import torch
import torch.nn as nn

# dice
Dice = torch.Tensor([0]).float().cuda()
input = torch.eye(3)
target = torch.zeros(3,3)
smooth = 0.01
# print(target)

intersect=(input*target).sum()
union = torch.sum(input) + torch.sum(target)
Dice=(2*intersect+smooth)/(union+smooth)
dice_loss=1-Dice
print(dice_loss)
```
