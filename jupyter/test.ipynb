```python
# gamma测试
import cv2
import numpy as np

def gamma_trans(img,gamma):
    # 先归一化到1，之后利用gamma作为指数求出新值，再还原
    gamma_table=[np.power(x/255.0,gamma)*255.0 for x in range(256)]
    gamma_table=np.round(np.array(gamma_table)).astype(np.uint8)

    # 用opencv的查表函数实现该映射
    return cv2.LUT(img,gamma_table)

img = cv2.imread("1.jpg")
cv2.imshow("img",img)
img_corrected = gamma_trans(img,0.5)
cv2.imshow("img_corrected",img_corrected)
cv2.waitKey(-1)
```

```python
# Leetcode024
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def swapPairs(self, head):
        dummy = ListNode(0,head)
        pCurrent = dummy
        while(pCurrent.next and pCurrent.next.next):
            node1 = pCurrent.next
            node2 = pCurrent.next.next
            node1.next = node2.next
            pCurrent.next = node2
            node2.next = node1
            pCurrent = pCurrent.next.next
        return dummy.next


```

```python
class ProgressBar(object):
    def __init__(self,mode,epoch=None,total_epoch=None,total=None,current=None,width=50,symbol=">"):
        self.mode = mode
        self.epoch = epoch
        self.total_epoch = total_epoch
        self.total = total
        self.current = current
        self.width = width
        self.symbol = symbol

    def __call__(self):
        percent = self.current/float(self.total)
        size = int(percent*self.width)
        bar = "["+self.symbol*size+" "*(self.width-size)+"]"
        args = {
            "percent":percent*100,
            "current":self.current,
            "total":self.total,
            "bar":bar,
            "mode":self.mode,
            "epoch":self.epoch+1,
            "epochs":self.total_epoch,
        }
        message = "\033[1;32;40m%(mode)s Epoch: %(epoch)d/%(epochs)d %(bar)s\033[0m [Current: LOSS=0.1 Top1=0.9] %(current)d/%(total)d \033[1;32;40m[ %(percent)3d%% ]"%args
        print("\r"+message,end="")
    
    def done(self):
        self.current = self.total
        print("")
        # with open("./logs/%s.txt" % self.model_name, "a") as f:
        #     print(self.write_message, file=f)


from time import sleep
progress = ProgressBar("Train",total_epoch=10)
for i in range(10):
    progress.total = 50
    progress.epoch = i
    for x in range(50):
        progress.current = x
        progress()
        sleep(0.1)
    progress.done()




        
```
